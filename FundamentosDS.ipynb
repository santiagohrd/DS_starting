```python
 # dos gráficas en una usando plt y sns
# plt.hist creamos un histograma de densidad. El argumento density=True hace que en lugar de mostrar cantidades absolutas, muestre proporciones o densidades.
# bins=50 significa que se crearán 50 barras en el histograma.
plt.hist(df_train['SalePrice'], density=True, bins=50)

#sns.kdeplot() agregamos encima la gráfica de KDE
sns.kdeplot(df_train['SalePrice'])
```

```python
cols_to_none = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu',
                'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
                'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1',
                'BsmtFinType2', 'MasVnrType']
df_train[cols_to_none] = df_train[cols_to_none].fillna('None')
```

Atributes that have most missing values

```python
total_missing = df_train.isnull().sum().sort_values(ascending=False)
percent_missing = (df_train.isnull().sum() / df_train.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([total_missing, percent_missing], axis=1, keys=['Total', 'Percent'])
missing_data = missing_data[missing_data['Total'] > 0].head(15)
print(missing_data)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=missing_data.index, y=missing_data['Percent'])
plt.xticks(rotation=90)
plt.title('Top 15 Features with Missing Values (by Percentage)')
plt.ylabel('Percent Missing')
plt.show()
```

```python
var = 'GrLivArea'
```

# **What is the correlation between sale price and the size of the basement?**

```python
data.plot.scatter(x=var, y='SalePrice', ylim=(0,800000));
```

Here we can see that the more big is the size of the basement the price of the house increase. Also we see few outliers that can be ignored or deleted from the DB to avoid the slant in the report

# **House quality vs price**

```python
OverallQual = 'OverallQual'
```

```python
data_overallqual = pd.concat([df_train['SalePrice'], df_train[OverallQual]], axis=1)
```

```python
f, ax = plt.subplots(figsize=(8,6))
fig = sns.boxplot(x=OverallQual, y='SalePrice', data = data_overallqual)
fig.axis(ymin=0, ymax=800000)
```

The quality of the house influences the final price, where we see that the max price increases.

# **Sale price vs construction year**

```python
yearbuilt = 'YearBuilt'
```

```python
dataYear = pd.concat([df_train['SalePrice'], df_train[yearbuilt]], axis=1)
```

```python
f, ax = plt.subplots(figsize=(20,8))
fig = sns.boxplot(x=yearbuilt, y='SalePrice', data=dataYear)
fig.axis(ymin=0, ymax=800000)
plt.xticks(rotation=90)
```

Sale price is correlated with how new is the house.

Confirming the correlation

```python
numeric_df = df_train.select_dtypes(include=np.number) #Incluye los valores númericos solamente
corrmat = numeric_df.corr()
```

All perfect correlations are those with clear color. In this case we can confirm that the quality, liveArea, basement, and yearBuilt vs salePrice are correlated.

```python
f, ax = plt.subplots(figsize=(12,9))
sns.heatmap(corrmat, vmax=8, square=True)
```

```python
k=10
```

```python
# Variables more related with the price
cols = corrmat.nlargest(k, 'SalePrice')['SalePrice'].index
```

```python
cols
```

```python
cm = np.corrcoef(df_train[cols].values.T)
```

```python
cm
```

```python
# Graphing the correlation
sns.set(font_scale=1.25)
```

```python
heatmap = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size':10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

Here we can see that:
* Quality, living area, area, basement are strongly correlated with the salePrice
* The variables of garage are correlated
* Basement and 1stFloor are like twins
* The year of construction are smothly correlated with salePrice
